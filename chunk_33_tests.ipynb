```python
df = pd.read_csv("data/chunk_33_prod.csv",index_col=None)
print len(df)
df.head(2)
```

```python
dfs = df[['Semana','Canal_ID','Agencia_ID','Ruta_SAK','Cliente_ID','Producto_ID','Demanda_uni_equil']]
```

```python
dfs['count'] = 1
summed = dfs.groupby(['Semana','Cliente_ID','Producto_ID']).sum()
```

```python
#this query gives all client product combos that occur twice in a given week
dups1 = summed.query('count == 2')
print len(dups1)
dups1
```

```python
#this query gives all client product combos that occur more than twice in a given week - watch out for product 53!!
dups1 = summed.query('count > 2')
print len(dups1)
dups1
```
